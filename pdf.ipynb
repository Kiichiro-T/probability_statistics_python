```
import numpy as np
from scipy import stats

%matplotlib inline
import matplotlib.pyplot as plt

!pip install japanize-matplotlib
import japanize_matplotlib
```

```
norm_rvs = stats.norm.rvs(loc=0, scale=20, size=10000, random_state=0)

plt.hist(norm_rvs, bins=10, alpha=0.3)
plt.xlabel('階級', fontsize=15)
plt.ylabel('度数', fontsize=15)
plt.show()
```

```
X = np.arange(start=-4, stop=4, step=0.1)
x = 2

y = stats.norm.pdf(x=x, loc=0, scale=1.0)
norm_pdf = stats.norm.pdf(x=X, loc=0, scale=1.0)

plt.plot(X, norm_pdf)
plt.plot(x, y, 'bo') # 青色ドット

plt.vlines(x, 0, y, linestyles='dashed')
plt.hlines(y, -4, x, linestyles='dashed')

plt.xlabel('X', fontsize=15)
plt.ylabel('pdf', fontsize=15)
plt.show()
```

```
p = 0.4
n = 5
k = np.arange(0, 5)

binom_pmf = stats.binom.pmf(k, n, p)

plt.plot(k, binom_pmf, 'bo', ms=8)
plt.vlines(k, 0, binom_pmf, colors='b', lw=3, alpha=0.5)
plt.xticks(k)

plt.xlabel('X', fontsize=15)
plt.ylabel('pmf', fontsize=15)
plt.show()
```

```
X = np.linspace(start=1, stop=7, num=100)

norm_cdf = stats.norm.cdf(x=X, loc=4, scale=0.8)
under_5 = stats.norm.cdf(x=5, loc=4, scale=0.8)

plt.plot(X, norm_cdf)
plt.plot(5, under_5, 'bo')

plt.vlines(5, 0.0, under_5, lw=2, linestyles='dashed')
plt.hlines(under_5, 1.0, 5, lw=2, linestyles='dashed')

plt.xlabel('X', fontsize=15)
plt.ylabel('cdf', fontsize=15)

plt.show()
```

```
X = np.arange(start=1, stop=7, step=0.1)

norm_pdf = stats.norm.pdf(x=X, loc=4, scale=0.8)

lower, upper = stats.norm.interval(alpha=0.95, loc=4, scale=0.8)
print('信頼区間95％の下限：', lower)
print('信頼区間95％の上限：', upper)

v_lower = stats.norm.pdf(x=lower, loc=4, scale=0.8)
v_upper = stats.norm.pdf(x=upper, loc=4, scale=0.8)

plt.plot(X, norm_pdf)
plt.plot(lower, 0.0, 'k|')
plt.plot(upper, 0.0, 'k|')

plt.vlines(lower, 0.0, v_lower, lw=0.8)
plt.vlines(upper, 0.0, v_upper, lw=0.8)

plt.xlabel("X", fontsize=13)
plt.ylabel("pdf", fontsize=13)

plt.show()
```
